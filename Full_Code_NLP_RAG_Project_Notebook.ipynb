## Question Answering using LLM

#### Downloading and Loading the model

```python
# Download the Mistral 7B GGUF model from Hugging Face Hub
model_path = hf_hub_download(
    repo_id="TheBloke/Mistral-7B-Instruct-v0.1-GGUF",
    filename="mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    repo_type="model"
)

# Load the model using llama-cpp-python
llm = Llama(
    model_path=model_path,
    n_ctx=4096,        # Context window
    n_threads=4,       # CPU threads
    n_gpu_layers=0     # Set > 0 for GPU acceleration
)
print("Model loaded successfully!")
```

#### Response

```python
def response(query,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    model_output = llm(
      prompt=query,
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      top_k=top_k
    )

    return model_output['choices'][0]['text']
```

## Data Preparation for RAG

### Loading the Data

```python
# uncomment and run the following lines for Google Colab
# from google.colab import drive
# drive.mount('/content/drive')
merck_pdf_path = "/content/medical_diagnosis_manual.pdf"
pdf_loader = PyMuPDFLoader(merck_pdf_path)
merck = pdf_loader.load()
```

### Data Overview

### Data Chunking

```python
# Split documents into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=100,
    length_function=len
)

chunks = text_splitter.split_documents(merck)
print(f"Total chunks created: {len(chunks)}")
print(f"\nSample chunk (first):\n{chunks[0].page_content[:300]}")
print(f"\nAverage chunk size: {sum(len(c.page_content) for c in chunks) // len(chunks)} characters")
```

### Embedding

```python
#Initialize SentenceTransformer embeddings
embedding_function = SentenceTransformerEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

# Test embedding on a sample text
sample_embedding = embedding_function.embed_query("What is sepsis?")
print(f"Embedding model: all-MiniLM-L6-v2")
print(f"Embedding dimensions: {len(sample_embedding)}")
print(f"Sample embedding (first 5 values): {sample_embedding[:5]}")
```

### Vector Database

```python
#Using the ChromaDB vector store
vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embedding_function,
    persist_directory="./merck_chroma_db",
    collection_name="merck_manual"
)

print(f"Vector store created with {vectorstore._collection.count()} documents")
print("ChromaDB persisted to ./merck_chroma_db")
```

### Retriever

```python
# Create retriever from vector store
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}  # Retrieve top-3 most relevant chunks
)

# Test retriever
test_docs = retriever.get_relevant_documents("sepsis management protocol")
print(f"Test retrieval - found {len(test_docs)} relevant chunks:")
for i, doc in enumerate(test_docs):
    print(f"\nChunk {i+1} (page {doc.metadata.get('page', 'N/A')}):")
    print(doc.page_content[:200] + "...")
```

### System and User Prompt Template

```python
# System message for the medical QnA assistant
qna_system_message = """You are a highly experienced medical professional and clinical reference specialist with deep expertise in the Merck Manual of Diagnosis and Therapy, 19th Edition.

Your role is to provide accurate, comprehensive, and evidence-based medical information based ONLY on the provided context from the Merck Manual.

Guidelines:
- Answer based strictly on the provided context; do not fabricate information
- Structure your response clearly with sections where appropriate
- Include dosages, protocols, and procedures as specified in the context
- Highlight critical warnings and contraindications
- If the context does not contain sufficient information, state this clearly"""

# User message template with placeholders for context and question
qna_user_message_template = """Context from Merck Manual:
{context}

Medical Question: {question}

Please provide a comprehensive answer based on the above context."""

print("System message and user template defined successfully.")
print(f"System message length: {len(qna_system_message)} characters")
```

### Response Function

```python
def generate_rag_response(user_input, k=3, max_tokens=256, temperature=0, top_p=0.95, top_k=50):
    global qna_system_message, qna_user_message_template

    # Step 1: Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input, k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Step 2: Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    # Step 3: Format the user message with context and question
    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    # Step 4: Combine system message and user message into prompt
    prompt = f"""[INST]{qna_system_message}

{user_message}[/INST]"""

    # Step 5: Generate the response using the LLM
    try:
        response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k
        )
        answer = response['choices'][0]['text']
        return answer, context_for_query, relevant_document_chunks
    except Exception as e:
        return f"Error generating response: {e}", "", []

print("RAG response function defined successfully.")
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message  = """You are an expert evaluator assessing the groundedness of AI-generated medical answers.

Groundedness measures whether the answer is supported by and consistent with the provided context, without hallucinating facts not present in the context.

Rate on a scale of 1-5:
1 - Answer contains significant information NOT in the context (hallucination)
2 - Answer partially uses context but adds unsupported claims
3 - Answer mostly uses context with minor unsupported additions
4 - Answer is well-grounded in context with minimal deviation
5 - Answer is entirely and accurately grounded in the provided context

Respond with a score of 1-5 and include a brief justification"""
```

```python
relevance_rater_system_message = """You are an expert evaluator assessing the relevance of AI-generated medical answers to the questions asked.

Relevance measures whether the answer directly addresses the specific question asked, without going off-topic.

Rate on a scale of 1-5:
1 - Answer is completely irrelevant to the question
2 - Answer is mostly irrelevant with minimal connection
3 - Answer partially addresses the question
4 - Answer is mostly relevant with minor gaps
5 - Answer directly and comprehensively addresses the question

Respond with a score of 1-5 and include a brief justification"""
```

```python
user_message_template = """Question: {question}

Retrieved Context:
{context}

Generated Answer:
{answer}"""
```

```python
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```
